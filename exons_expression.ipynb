```python
# uniwue tissue names in the df
names = splice_variants.columns.values
groups = []
for i in range(len(names)):
    curr_group = str(names[i]).split('_')[1:]
    groups.append(curr_group)
    
groups = np.unique(groups).tolist()
tissues = []
for curr in groups:
    tissues.append('_'.join(curr))
tissues.remove(tissues[-1])  #id
```

```python
splice_variants = splice_variants.T
splice_variants.columns = splice_variants.columns[:-1].tolist() + ['Complete gene']
splice_variants
```

```python
df1 = splice_variants.melt(var_name = 'Genes', value_name = 'Expression', ignore_index=False)
df1
```

```python
for i in tissues:
    df2 = df1.loc[df1.index.str.endswith(i)].reset_index()
    df2 = df2.sort_values("Expression")
    
    print(i)
    for j in df2.columns[1:]:
        fig, ax = plt.subplots(figsize=(15, 10))
        sns.barplot(data=df2, x="index", y='Expression', hue='Genes')
        
        ax.set_xticklabels(ax.xaxis.get_ticklabels(), rotation=90)
        ax.set_xlabel('Cell line')
        plt.tight_layout()
        
        fig.savefig(f'{i.split(".")[0]}.jpg', facecolor='white')
        plt.close()
```
