```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
splice_variants = pd.read_csv('ADGRL1_transcripts.tsv', sep='\t', index_col=0)
gene = pd.read_csv('ADGRL1_gene.tsv', sep='\t', index_col=0)
```

```python
splice_variants # экспрессия по сплайс-вариантам
```

```python
splice_variants=splice_variants.reset_index(drop=True)
splice_variants
```

```python
gene = gene.reset_index(drop=True)
gene
```

```python
gene.rename(columns={'transcript_ids':'transcript_id'}, inplace=True)
gene
```

```python
dfs = [splice_variants, gene]
splice_variants = pd.concat(dfs)
splice_variants
```

```python
splice_variants # сплайс варианты + полный ген
```

```python
splice_variants=splice_variants.set_index('transcript_id')
splice_variants
```

```python
splice_variants.to_csv('result_concat_splices.csv')
```

```python
names = splice_variants.columns.values
groups = []
for i in range(len(names)):
    curr_group = str(names[i]).split('_')[1:]
    groups.append(curr_group)
```

```python
groups = np.unique(groups)
groups = groups.tolist()
tissues = []
for curr in groups:
    tissues.append('_'.join(curr))
tissues.remove(tissues[-1])  #id
```

```python
splice_variants = splice_variants.T
splice_variants.columns = splice_variants.columns[:-1].tolist() + ['Complete gene']
splice_variants
```

```python
for i in tissues:
    df = splice_variants.loc[splice_variants.index.str.endswith(i)].reset_index()
    print(i)
    for j in df.columns[1:]:
        fig, ax = plt.subplots(figsize=(15, 15))
        sns.barplot(data=df, y=j, x='index', ax=ax)
        ax.set_xticklabels(ax.xaxis.get_ticklabels(), rotation=90)
        ax.set_xlabel('Cell line')
        plt.tight_layout()
        fig.savefig(f'pictures_tissue/{i.split(".")[0]}_{j.split(".")[0]}.png', facecolor='white')
#         plt.show()
        plt.close()
```
