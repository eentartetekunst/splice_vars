```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
splice_variants = pd.read_csv('ADGRL1_transcripts.tsv', sep='\t', index_col=0)
gene = pd.read_csv('ADGRL1_gene.tsv', sep='\t', index_col=0)
gene
```

```python
gene = gene.reset_index(drop=True)
gene
```

```python
gene.rename(columns={'transcript_ids':'transcript_id'}, inplace=True)
gene
```

```python
dfs = [splice_variants, gene]
splice_variants = pd.concat(dfs)
splice_variants
```

```python
names = splice_variants.columns.values
groups = []
for i in range(len(names)):
    curr_group = str(names[i]).split('_')[1:]
    groups.append(curr_group)
```

```python
groups = np.unique(groups)
groups = groups.tolist()
tissues = []
for curr in groups:
    tissues.append('_'.join(curr))
tissues.remove(tissues[-1])  #id
```

```python
scale = 0.5

for i in tissues:
    
    df = splice_variants.loc[splice_variants.index.str.endswith(i)].reset_index()
    tidy = df.melt(id_vars='index').rename({'index': 'index', 'variable': 'gene', 'value': 'value'}, axis=1)
    tmp = tidy.groupby('index').sum()
    tidy = pd.merge(tidy, tmp, on='index').sort_values(by='value_y')
    height = tmp.shape[0]*scale if tmp.shape[0] > 3 else 5 
    fig, ax = plt.subplots(figsize=(15, height))
    sns.barplot(data=tidy, y='index', x='value_x', hue='gene', ax=ax)
    ax.set_xlabel('expression') 
    ax.set_ylabel('cell line')
        
    plt.tight_layout()
    fig.savefig(f'graphs/{i.split(".")[0]}.png', facecolor='white')
    plt.close()
```
